### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
items=soup.find(class_="part_list_2").find_all('h3')
for item in items:
    print(item.find(class_="date").text, item.find('em').text,item.find_all('a')[-1].text)
   
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
cat_list = []
for cat in items:
    cat_list.append(cat.find('em').text)
cat_type = set(cat_list)
list=[]
for type in cat_type:
    for item in items:
        if item.find('em').text==type:
            list.append(item.find('em').text)
            list.append(item.find(class_="date").text)
            list.append(item.find('a').text)
print(list)
```

### ③ 哪一個類別的文章最多

```python
cat_dict = {}
for key in cat_list:
    cat_dict[key] = cat_dict.get(key, 0) + 1
print('類別的文章最多 :',max(cat_dict),'數量 :',cat_dict[key])
```
