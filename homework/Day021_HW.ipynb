## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import time

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

date_time = datetime.now() - timedelta(1)
date_time = date_time.strftime("%Y/%m/%d")
#print("取得昨天時間：", date_time)

while True:
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight),2000;")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    #print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)
    if date_time in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        #print('==== STOP ====')
        break
        
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
for d in soup.find_all(class_="part_list_2")[-1].find_all('h3'):
    if date_time not in d.find(class_="date").text:      
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime, timedelta
import time

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

date_time = datetime.now() - timedelta(1)
date_time = date_time.strftime("%Y/%m/%d")
#print("取得昨天時間：", date_time)

while True:
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight),2000;")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    #print(soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text)
    if date_time in soup.find_all(class_="part_list_2")[-1].find_all('h3')[-1].find(class_="date").text:
        #print('==== STOP ====')
        break

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
for d in soup.find_all(class_="part_list_2")[-1].find_all('h3'):
    if date_time not in d.find(class_="date").text:
        if (int(d.span.text[11:13])>=15 and int(d.span.text[11:13])<=17):
            print(d.find(class_="date").text, d.find_all('a')[-1].text)
       
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select
from datetime import timedelta, datetime
import time

three_age = datetime.today() + timedelta(-3)
fotmated_day = three_age.strftime('%Y/%m/%d')

browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")

selectSite = Select(browser.find_element_by_id("selD"))
selectSite.select_by_value(fotmated_day[8:10])
browser.find_element_by_id('button').click()

# 設定判斷控制器，僅取得當日擷取範圍
keepGoing = True
while keepGoing:    
    time.sleep(2)
    browser.execute_script("window.scrollBy(0,10000)")
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")

    div_find_h3 = soup.find(class_="part_list_2").find_all('h3')
    for i in div_find_h3:
        if (str(i.span.text[:10]) == fotmated_day):
            keepGoing = True
        else:
            keepGoing = False                        

# 解析相關網頁內容
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
for i in div_find_h3:
    if (int(i.span.text[11:13])>=15 and int(i.span.text[11:13])<=17):
        print(i.span.text,i.em.text,i.a.text,i.span.text[11:13])
print('======3天前新聞擷取完畢======')                        
browser.quit()
```
